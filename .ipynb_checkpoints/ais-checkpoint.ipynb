### PySpark merge

```python
# Check if 'spark' exists in locals()
if 'spark' in locals():
    # If 'spark' exists, print the current SparkSession
    print(f"Current SparkSession - {spark}")
else:
    # If 'spark' doesn't exist, create a new SparkSession
    spark = SparkSession.builder \
        .appName("AIS") \
        .config("spark.executor.memory", "8g") \
        .config("spark.driver.memory", "8g") \
        .getOrCreate()
    print(f"New SparkSession {spark} created.")
```

```python
# Carregue os DataFrames PySpark
df_agroFood_spark = spark.createDataFrame(df_agroFood)
df_foodEmissions_spark = spark.createDataFrame(df_foodEmissions)
df_globalFood_spark = spark.createDataFrame(df_globalFood)
df_fao_spark = spark.createDataFrame(df_fao)

# Inicie o cronômetro
start_time = time.time()

# Realize os merges usando PySpark
df1_spark = df_agroFood_spark.join(df_foodEmissions_spark, ['Country', 'Year'], 'inner')
df2_spark = df1_spark.join(df_globalFood_spark, ['Country', 'Year'], 'inner')
dfinal_spark = df2_spark.join(df_fao_spark, ['Country', 'Year'], 'inner')

# Pare o cronômetro
elapsed_time_pyspark = time.time() - start_time

#dfinal_spark.show()
print(f"Tempo decorrido para o merge (PySpark): {elapsed_time_pyspark:.6f} segundos")
```

```python
def insert_batch(i, batch_df, session, prepared, queue):
    row_count = 0
    start_time = time.time()

    # Insert data into Cassandra for the current batch
    for row in batch_df:
        session.execute(prepared, (
            row['index'], row['country'], row['year'], row['savanna fires'], row['forest fires'], row['crop residues'], 
            row['rice cultivation'], row['drained organic soils (co2)'], row['pesticides manufacturing'], 
            row['food transport'], row['forestland'], row['net forest conversion'], row['food household consumption'], 
            row['food retail'], row['on-farm electricity use'], row['food packaging'], 
            row['agrifood systems waste disposal'], row['food processing'], row['fertilizers manufacturing'], 
            row['ippu'], row['manure applied to soils'], row['manure left on pasture'], row['manure management'], 
            row['fires in organic soils'], row['fires in humid tropical forests'], row['on-farm energy use'], 
            row['rural population'], row['urban population'], row['emissions_food'], row['average temperature'], row['item'], 
            row['element'], row['unit'], row['food_emissions'], row['population'], row['production (t)'], 
            row['production__tonnes__per_capita'], row['production per capita (kg)'], row['yield (t/ha)'], 
            row['land use (ha)'], row['area_harvested__ha__per_capita'], row['land use per capita (m2)'], 
            row['imports (t)'], row['imports__tonnes__per_capita'], row['imports per capita (kg)'], 
            row['exports (t)'], row['exports__tonnes__per_capita'], row['exports per capita (kg)'], 
            row['domestic supply (t)'], row['domestic_supply__tonnes__per_capita'], 
            row['domestic supply per capita (kg)'], row['food (t)'], row['food__tonnes__per_capita'], 
            row['food per capita (kg)'], row['supply chain waste (t)'], 
            row['waste_in_supply_chain__tonnes__per_capita'], row['supply chain waste per capita (kg)'], 
            row['food supply (kg per capita per year)'], row['food supply (g per capita per day)'], 
            row['food supply (kcal per capita per day)'], row['food supply (protein g per capita per day)'], 
            row['food supply (fat g per capita per day)'], row['m49_code'], row['cpc_code'], row['commodity'], 
            row['loss_percentage'], row['loss_percentage_original'], row['activity'], row['food_supply_stage'], 
            row['method_data_collection'], row['url']
        ))
        row_count += 1

    end_time = time.time()
    duration = end_time - start_time
    queue.put((i, row_count, duration))

def main():
    print("Start time:", datetime.datetime.now().strftime("%H:%M:%S"))
    time_passed = 0
    #num_batches = ... # Define the number of batches
    #batch_size = ... # Define the batch size
    #df = ... # Assuming df is your DataFrame

    # Split the DataFrame into batches
    batches = [df[i:i + batch_size] for i in range(0, df.shape[0], batch_size)]

    # Create a queue to collect results from each process
    queue = Queue()

    # Create and start processes
    processes = []
    for i, batch_df in enumerate(batches):
        p = Process(target=insert_batch, args=(i, batch_df, session, prepared, queue))
        p.start()
        processes.append(p)

    # Wait for all processes to finish and collect results
    for p in processes:
        p.join()

    # Process results from the queue
    while not queue.empty():
        i, row_count, duration = queue.get()
        duration_str = "{:.2f}".format(duration / 60)
        print(f"Batch {i+1} inserted with {row_count} rows in {duration_str} minutes")
        time_passed += duration

    print("--------------------------")
    print("End time:", datetime.datetime.now().strftime("%H:%M:%S"))

if __name__ == "__main__":
    main()
```
